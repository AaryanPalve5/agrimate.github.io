## Stationarity Test

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def ad_test(dataset):
  dftest = adfuller(dataset, autolag = "AIC")
  print("1. ADF: ", dftest[0])
  print("2. P-value: ", dftest[1])
  print("3. Num of lags: ", dftest[2])
  print("4. Num Of Observations Used For ADF Regression:", dftest[3])
  print("5. Critical value: ")
  for key, val in dftest[4].items():
    print("\t", key, ":", val)
```

```python
ad_test(df2["Rainfall"])
```

## As P-value is less than 0.05, dataset is stationary

### Best parameters coming out to be (2, 0, 1)

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
print(df3.shape)
train = df3.iloc[:-90]
test = df3.iloc[-90:]
print(train.shape,test.shape)
```

```python
model=ARIMA(train['Rainfall'],order=(2,0,1))
model=model.fit()
model.summary()
```

```python
start=len(train)
end=len(train)+len(test)-1
pred=model.predict(start=start,end=end,typ='levels').rename('ARIMA Predictions')
pred.plot(legend=True)
test['Rainfall'].plot(legend=True)
```

```python
print(pred)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
test['Rainfall'].mean()
rmse=sqrt(mean_squared_error(pred,test['Rainfall']))
print(rmse)
```

## Error is quite high!

## Introducing Seasonality component as model was unable to capture seasonality

```python
model1 = auto_arima(y = train.Rainfall, m = 12)
```

```python
predictions = pd.Series(model1.predict(n_periods = len(test)))
predictions.index = test.index
predictions
```

```python
test["Rainfall"].plot(figsize = (12, 8), legend = True)
predictions.plot(legend = True)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
test['Rainfall'].mean()
rmse=sqrt(mean_squared_error(predictions,test['Rainfall']))
print(rmse)
```

## Noticable decrease in error
